# read the data

this comes from the `data wrangling one beta farm.ipynb` notebook

read the dates from the txt file

```python
# calculate a rolling median with a 25 day window, minimum of 3 valid points

output_fname = 'old_redbank_mob-buffer_smoothed.nc'
data_rolling = xr.open_dataset(output_fname)
data_rolling = data_rolling.squeeze()
data_rolling
```

```python
#aggregate to monthly
data_monthly = data_rolling.resample(time="M").mean(dim="time")
data_monthly
```

```python
# even aggregated to monthly the data has gaps, use a simple linear interpolation to fill them
data_monthly = data_monthly.interpolate_na(dim='time')
data_monthly
```

```python
# put the three bands into a stacked numpy array
data = np.vstack(
    [data_monthly.sel(time=slice('2018','2022'))['bare'].to_numpy(),
     data_monthly.sel(time=slice('2018','2022'))['green'].to_numpy(),
     data_monthly.sel(time=slice('2018','2022'))['non-green'].to_numpy()])  
dates=data_monthly.sel(time=slice('2018','2022')).time
data_reshape = data.reshape(data.shape[0], -1) 

data.shape, data_reshape.shape
```

```python
# reshape the data to 2D and identify pixels with good data
# run a robust scaler
data_reshape = data.reshape(data.shape[0], -1) 
good_data = np.all(np.isfinite(data_reshape),axis=0)
X_Robust = RobustScaler().fit(data_reshape[:,good_data].T).transform(data_reshape[:,good_data].T)
X_Robust.shape
```

```python
# run a PCA to reduce the dimensionality. Retain 90% of the variability
pca = PCA(n_components=0.9)
X_pca = pca.fit_transform(X_Robust)
#X_pca = (X_pca*100).round().astype(int)
X_pca_reshape.shape, pca.explained_variance_ratio_, np.cumsum(pca.explained_variance_ratio_)
```

```python
# run a kmeans clustering 

from sklearn.cluster import KMeans
n_clusters = 20 

clusterer = KMeans(n_clusters=n_clusters, random_state=10)
cluster_labels = clusterer.fit_predict(X_pca)+1
 
```

```python
cluster_labels_reshape = np.zeros((data_reshape.shape[1])).astype(int)
cluster_labels_reshape[good_data] = cluster_labels 
cluster_labels_reshape = cluster_labels_reshape.T.reshape(data.shape[1],data.shape[2])
cluster_labels_reshape.shape, data_reshape.shape
```

```python
# put  cluster_labes back in the xarray, makes it easier for the analyses later on 

cluster_labels_da= xr.full_like(data_monthly['bare'].isel(time=0).astype(int).squeeze().drop_vars(['time','quantile']), fill_value=0)
cluster_labels_da.name = 'cluster_labels'
merged =  xr.merge([data_monthly, cluster_labels_da]) 
merged['cluster_labels'] = (['y','x'],cluster_labels_reshape)
merged
```

```python
# show me a map of the clusters
merged['cluster_labels'].plot(cmap='tab20')
```

```python
# save the clusters to a tif file so it can be opened in QGis or other software

merged['cluster_labels'].rio.to_raster('old_redbank_clustered.tif', tiled=True)
```

```python
# convert to pandas , makes it easier to plot with seaborn
merged_df=merged.to_dataframe().reset_index()
merged_df = merged_df.loc[merged_df['cluster_labels'] != 0]
merged_df = merged_df.groupby(['time','cluster_labels']).mean().reset_index()
#merged_df['cluster_labels']  = merged_df['cluster_labels'].astype(str)
merged_df['cluster_labels'] = merged_df['cluster_labels'].apply(lambda x: f"{x:02d}")
merged_df
```

```python
data_sns = merged_df.groupby(['time','cluster_labels']).mean().melt(value_vars=['bare', 'green', 'non-green'], ignore_index=False).reset_index()
data_sns
```

```python
data_sns['time'].dt.month
```

```python
sns.relplot(
    data=data_sns,
    x="time", y="value", hue="cluster_labels", row='variable', aspect=3, height=3,
    kind="line", palette="tab20"
)
```

```python
# same as above ut split by year
sns.relplot(
    data=data_sns,
    x=data_sns['time'].dt.month, y="value", hue="cluster_labels", row='variable', col = data_sns['time'].dt.year , aspect=1, height=2,
    kind="line", palette="tab20"
)
```
